# **8. Data Exploration**

In this section, the dataset was explored to understand its structure and features. Functions like `head()`, `info()`, and `describe()` were used to examine sample data, data types, and statistical summaries. This helped identify missing values and understand key columns such as type, release year, genres, and ratings before proceeding to data cleaning and analysis.

```python
# Load dataset files
titles = pd.read_csv('titles.csv')
credits = pd.read_csv('credits.csv')

# Display first few rows
titles.head()
```

```python
# Check column names
titles.columns
# Get dataset info
titles.info()
# Summary statistics
titles.describe()
```

**4. Data Cleaning**

Missing values are handled to ensure accurate analysis. Numerical and categorical columns are treated separately.

```python
# Check missing values
titles.isnull().sum()

# Separate numeric and categorical columns
num_cols = titles.select_dtypes(include=['float64', 'int64']).columns
cat_cols = titles.select_dtypes(include=['object']).columns

# Fill numeric values with median
titles[num_cols] = titles[num_cols].fillna(titles[num_cols].median())

# Fill categorical values with 'Unknown'
titles[cat_cols] = titles[cat_cols].fillna('Unknown')
```

# **10. Data Analysis and Visualizations**

This section includes various visualizations to analyze trends, distributions, and relationships in the dataset.

This graph shows the distribution of movies and TV shows on Amazon Prime.

1. This visualization shows the distribution of movies and TV shows available on Amazon Prime.

```python
# Plot distribution of movies vs TV shows
sns.countplot(x='type', data=titles)
plt.title("Movies vs TV Shows")
plt.show()
```

2. This pie chart represents the percentage distribution of movies and TV shows on the platform.

```python
# Pie chart to show proportion of content types
titles['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Content Type Distribution")
plt.ylabel('')
plt.show()
```

3. This graph shows the most common genres available on Amazon Prime after cleaning and splitting the genre data.

```python
# Clean and split genres for accurate analysis
titles['genres'] = titles['genres'].astype(str)
titles['genres'] = titles['genres'].str.replace(r"[\[\]']", "", regex=True)
titles['genres'] = titles['genres'].str.split(',')

genres_exploded = titles.explode('genres')
genres_exploded['genres'] = genres_exploded['genres'].str.strip()

# Plot top genres
genres_exploded['genres'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.xticks(rotation=45)
plt.show()
```

4. This pie chart highlights the contribution of the top genres to the overall content.

```python
# Pie chart of top genres distribution
genres_exploded['genres'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Top Genres Distribution")
plt.ylabel('')
plt.show()
```

5. This visualization shows how average IMDb ratings vary across different genres.

```python
# Compare average ratings across genres
genres_exploded.groupby('genres')['imdb_score'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Average IMDb Score by Genre")
plt.xticks(rotation=45)
plt.show()
```

6. This graph shows how the number of movies and TV shows released has changed over time.

```python
# Plot growth of content over years
titles['release_year'].value_counts().sort_index().plot()
plt.title("Content Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

7. This line chart shows the trend of movies and TV shows over time, providing a clearer view compared to bar charts.

```python
# Line plot for better visualization of trends over time
data = titles.groupby(['release_year','type']).size().unstack()

plt.figure(figsize=(10,5))
data.plot()

plt.title("Movies vs Shows Over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

8. This graph groups content by decade to show long-term trends in content production.

```python
# Group release years into decades for trend analysis
titles['decade'] = (titles['release_year']//10)*10
titles['decade'].value_counts().sort_index().plot(kind='bar')
plt.title("Content by Decade")
plt.show()
```

9. This graph shows how the average IMDb rating has changed over the years.

```python
# Track changes in average ratings over time
titles.groupby('release_year')['imdb_score'].mean().plot()
plt.title("Average IMDb Score Over Years")
plt.show()
```

10. This histogram shows the distribution of IMDb scores across all titles.

```python
# Distribution of IMDb scores
sns.histplot(titles['imdb_score'], bins=20)
plt.title("IMDb Score Distribution")
plt.show()
```

11. This boxplot shows the distribution of IMDb scores. Most content is concentrated between ratings of 5 and 7, with a median around 6. The presence of multiple outliers on both lower and higher ends indicates variability in content quality.

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=titles['imdb_score'], showfliers=False)

plt.title("IMDb Score Distribution (Boxplot)")
plt.xlabel("IMDb Score")

plt.tight_layout()
plt.show()
```

12. This histogram shows the distribution of TMDB popularity after applying a log transformation to reduce skewness. It helps in better visualizing how most titles have low to moderate popularity, while only a few have very high popularity values.

```python
plt.figure(figsize=(8,5))

sns.histplot(np.log1p(titles['tmdb_popularity']), bins=30)

plt.title("TMDB Popularity Distribution (Log Scale)")
plt.xlabel("Log Popularity")
plt.show()
```

13. This scatter plot shows the relationship between IMDb ratings and popularity.

```python
# Scatter plot to compare rating vs popularity
sns.scatterplot(x='imdb_score', y='tmdb_popularity', data=titles)
plt.title("Popularity vs Rating")
plt.show()
```

14. This heatmap shows the correlation between numerical features such as ratings and popularity.

```python
# Correlation matrix for numerical features
sns.heatmap(titles[['imdb_score','tmdb_score','tmdb_popularity']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

15. This chart shows the most popular titles based on TMDB popularity scores.

```python
# Plot top 10 titles based on TMDB popularity using a horizontal bar chart
top_pop = titles.sort_values(by='tmdb_popularity', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_pop['title'], top_pop['tmdb_popularity'])

plt.title("Top 10 Most Popular Titles")
plt.xlabel("Popularity")
plt.gca().invert_yaxis()

plt.show()
```

16. This chart shows the top-rated titles based on IMDb scores with adjusted scale for better comparison.

```python
# Plot top 10 titles by IMDb score and set x-axis limits for clearer comparison
top_rated = titles.sort_values(by='imdb_score', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_rated['title'], top_rated['imdb_score'])

plt.title("Top Rated Titles")
plt.xlabel("IMDb Score")
plt.xlim(8.5, 10)  # zoom in

plt.gca().invert_yaxis()
plt.show()
```

17. This graph shows which countries produce the most content available on Amazon Prime.

```python
# Analyze content production by country
titles['production_countries'].astype(str)\
.str.replace(r"[\[\]']", "", regex=True)\
.str.split(',')\
.explode()\
.str.strip()\
.value_counts().head(10)\
.plot(kind='bar')

plt.title("Top Production Countries")
plt.xticks(rotation=45)
plt.show()
```
